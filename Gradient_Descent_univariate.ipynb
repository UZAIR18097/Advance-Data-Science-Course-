```
precision  = 0.00001 #when to stop the algorithm
x =2
alpha = 0.001
change_in_x = 1
iters = 0 #iteration number to count the number of iteration at end

while change_in_x > precision:
  prev_x = x #store the current value of x in prev_x
  x = x -alpha*(2*(x+5)) #Gradatient descent
  change_in_x = abs(x-prev_x)
  iters += 1
  print("Iterations", iters, "\nx value is",x)

print("The minima occurs at" ,x)


```
